For scATAC-seq analysis, STREAM can take as input a feature matrix, 

- e.g. k-mer feature matrix
- e.g. motif feature matrix

More details can be found https://github.com/pinellolab/STREAM_atac

````python
adata = st.read(file_name='./zscores_scaled.tsv.gz',experiment='atac-seq')
````

To get the **zscore file**, 

User can run the following codes within Jupyter Notebook (**Step[3] can be skipped**):
> Using k-mers as features:
```python
import stream_atac  
adata = stream_atac.preprocess_atac(file_count='./count_file.tsv.gz',file_region='./region_file.bed.gz',file_sample='./sample_file.tsv.gz',genome = 'hg19')
```

> Using motifs as features:
```python
import stream_atac  
adata = stream_atac.preprocess_atac(file_count='./count_file.tsv.gz',file_region='./region_file.bed.gz',file_sample='./sample_file.tsv.gz',genome = 'hg19',feature='motif')
```

or user can run the following codes in Terminal:
> Using k-mers as features:  
`stream_atac -c count_file.tsv.gz -s sample_file.tsv.gz -r region_file.bed.gz -g hg19`

> Using motifs as features:  
`stream_atac -c count_file.tsv.gz -s sample_file.tsv.gz -r region_file.bed.gz -g hg19 -f motif`

To load and use 10x Genomics single cell ATAC-seq data processed with Cell Ranger:

By default **factor_min_win=2.0**, adjusting it (between 1.0 and 2.0) can make smoother stream plot

````python
st.stream_plot(adata,root='S2',fig_legend_ncol=6,fig_size=(8,8),factor_min_win=1.8)
````

For stream plot with **'thin branches'**, log2 view of stream plot (by specifying **'flag_log_view=True'**) will help zoom in the thin branches with very few cells by log2 transformation and visualize the composition of cells on the thin branches  
**factor_zoomin**: suggested value 50~200

````python
st.stream_plot(adata,root='S2',fig_legend_ncol=6,fig_size=(8,8),factor_min_win=1.8,flag_log_view=True,factor_zoomin=50)
````

````python
st.stream_plot(adata,root='S2',fig_legend_ncol=6,fig_size=(8,8),factor_min_win=1.8,flag_log_view=True,factor_zoomin=100)
````
